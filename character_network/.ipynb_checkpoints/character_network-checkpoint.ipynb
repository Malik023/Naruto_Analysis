```python
from glob import glob
# Retrieving and sorting the dataset
substitles_paths = sorted(glob("C:/Users/malik/Downloads/Naruto_Subtitles/*.ass"))

```

```python
# Initializs empty list to store scripts and episode numbers
scripts = []
episode_num = []

# Iterate over each file within substitles_paths
for path in substitles_paths:

    # Open the file and read its contents
    with open(path,'r') as file:
        # Read all the lines within the file
        lines = file.readlines()
        # Skip the first 27 lines
        lines = lines[27:]

        # Extract relevant parts from each line and join them
        rows = [",".join(line.split(',')[9:]) for line in lines]
    # Format/clean the rows    
    rows = [line.replace("\\N",' ') for line in rows]
    # Combine all the cleaned rows into a signle script
    script = " ".join(rows)
    # Extract episode number from the file path
    episode = int(path.split('-')[1].split('.')[0].strip())
    # Append the script and episode number to their respective lists
    scripts.append(script)
    episode_num.append(episode)
```

```python
import pandas as pd
# Create a DataFrame from the lists of episode numbers and scripts
df = pd.DataFrame.from_dict({'episode':episode_num,'script':scripts})
```

```python
df['ners'] = df['script'].apply(get_ners)
```

```python
# Define the window size for considering previous entities
window = 10

# List to store entity relationships
entity_relationship = []

# Iterate through each row in the 'ners' column of the DataFrame
for row in df['ners']:
    previous_entities_in_window = []
    
    # Iterate through each sentence in the row
    for sentence in row:
        # Add the current sentence to the list of previous sentences within the window
        previous_entities_in_window.append(sentence)
        # Limit the number of previous sentences to the window size
        previous_entities_in_window = previous_entities_in_window[-window:]
        
        # Flatten the list of previous sentences
        previous_entities_flattened = sum(previous_entities_in_window, [])
        
        # Iterate through each entity in the current sentence
        for entity in sentence:            
            # Iterate through each entity in the flattened list of previous sentences
            for entity_in_window in previous_entities_flattened:
                # Ensure that the entities are different
                if entity != entity_in_window:
                    # Sort the pair of entities alphabetically to ensure consistency
                    entity_rel = sorted([entity, entity_in_window])
                    # Add the entity relationship to the list
                    entity_relationship.append(entity_rel)
```

```python
relationship_df = pd.DataFrame({'value':entity_relationship})
```

```python
# Create a new column 'source' by extracting the first entity from the 'value' column
relationship_df['source'] = relationship_df['value'].apply(lambda x: x[0])
# Create a new column 'target' by extracting the second entity from the 'value' column
relationship_df['target'] = relationship_df['value'].apply(lambda x: x[1])
```

```python
# Group the DataFrame by 'source' and 'target' columns and count the occurrences
relationship_df = relationship_df.groupby(['source','target']).count().reset_index()
```

```python
# Sort the values within the raltionship_df Dataframe
relationship_df = relationship_df.sort_values('value',ascending=False)
relationship_df.head()
```

```python
relationship_df = relationship_df.head(200)
```

```python
import networkx as nx
# Create a graph from the pandas DataFrame using the 'source' and 'target' columns as nodes
# and the 'value' column as edge attributes
G = nx.from_pandas_edgelist(relationship_df, 
                            source = "source", 
                            target = "target", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```
